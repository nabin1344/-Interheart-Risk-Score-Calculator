```python
import pandas as pd

# Create a sample DataFrame with a categorical column

data = {'BP': ['High', 'High', 'Low', 'Medium', 'High'],
        "BP_level":[150, 200, 120, 140, 155],
        'HTN': ["YES", "YES", "No", "Yes", "Yes"]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
# Formula: Predict 'survived' using 'age', 'sex', and 'pclass'
# C(sex) and C(pclass) are treated as categorical
model_formula = smf.logit("HTN ~ BP_level + C(BP) ", data=df)

# Fit the model
result = model_formula.fit()

# Print the summary
print(result.summary())

```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# ---------------------------------------------------
# 1. Create DataFrame
# ---------------------------------------------------
data = {
    'Sex': ['Female', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male', 'Female'],
    'BP_level': [150, 200, 120, 140, 155, 140, 120, 145, 160, 160],
    'Diabetes': ["Yes", "No", "No", "No", "No", "No", "No", "Yes", "Yes", "Yes"],
    'HTN': ["Yes", "Yes", "No", "Yes", "Yes", "Yes", "No", "No", "Yes", "Yes"],

}

df = pd.DataFrame(data)

print("\n===== ORIGINAL DATA =====")
print(df)

# ---------------------------------------------------
# 2. Convert Target Variable to Binary
# ---------------------------------------------------
df['HTN_binary'] = df['HTN'].map({"Yes": 1, "No": 0})
print("\n===== DATA WITH BINARY COLUMNS =====")
print(df)
print("-" * 40)

# ---------------------------------------------------
# 3. Logistic Regression Model
# ---------------------------------------------------
# C(BP) automatically creates dummy variables for BP categories
formula = "HTN_binary ~ BP_level + C(Sex)"

model = smf.logit(formula, data=df)

# Fit model
result = model.fit()

# ---------------------------------------------------
# 4. Output Results
# ---------------------------------------------------
print("\n===== LOGISTIC REGRESSION SUMMARY =====")
print(result.summary())

# ---------------------------------------------------
# 5. Odds Ratios
# ---------------------------------------------------
odds_ratios = np.exp(result.params)

print("\n===== ODDS RATIOS =====")
print(odds_ratios)
```

```python
#1. Model Fit

# Pseudo R-squared = 0.6205 → The model explains ~62% of the variation in hypertension.

# LLR p-value = 0.0226 → The model as a whole is statistically significant (p < 0.05).

# Observations = 10 → Very small sample; results should be interpreted cautiously.
```

```python
# 2. Coefficients and Odds Ratios

# | Predictor | Coef | P>|z| | Odds Ratio | Interpretation |
# |------------------|-----------|------|------------|----------------|
# | Intercept | -29.3420 | 0.160 | 1.81e-13 | Baseline odds for a Female with BP_level = 0 (not meaningful practically). |
# | C(Sex)[T.Male] | -2.7647 | 0.418 | 0.063 | Being Male reduces odds of HTN by ~94% compared to Female (1-0.063), but not statistically significant (p=0.418). |
# | BP_level | 0.2262 | 0.154 | 1.254 | Each 1 mmHg increase in BP_level increases odds of HTN by 25%, but not statistically significant (p=0.154). |

# . Interpretation in Words

# In this small sample, higher blood pressure (BP_level) appears to increase the likelihood of hypertension:
#  each 1 mmHg rise in BP_level increases the odds by about 25%. Males have lower odds of hypertension compared to females (~94% lower),
#  although this effect is not statistically significant. Overall, the model is significant, suggesting BP_level and
#   sex together have predictive value,
#   but individual predictors are not significant likely due to small sample size.
```

```python
# 2. Interpret OR in Percentage Terms

# If OR > 1 → predictor increases odds of outcome.

# If OR < 1 → predictor decreases odds of outcome.

# Formula:

# Percentage change in odds =( OR − 1 ) × 100

# Sex (Male)

# OR = 0.063

# Percentage change:

# (0.063−1)×100=−0.937×100=−93.7%
# ✅ Interpretation: Being Male reduces the odds of HTN by ~94% compared to Female.

# Negative value → decrease in odds.

# Positive value → increase in odds.
```
